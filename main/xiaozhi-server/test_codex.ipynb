```python
import os
import sys
import json
from pathlib import Path

sys.path.insert(0, r"C:\Users\yibinjiang\Documents\GitHub\xiaozhi-esp32-server\main\xiaozhi-server")

from core.providers.llm.codex_app_server.codex_app_server import LLMProvider

SESSION_ID = "s1"
SESSION_FILE = Path(r"C:\Users\yibinjiang\Documents\GitHub\xiaozhi-esp32-server\.codex_session.json")


def load_thread_id():
    if SESSION_FILE.exists():
        try:
            return json.loads(SESSION_FILE.read_text(encoding="utf-8")).get("thread_id")
        except Exception:
            return None
    return None


def save_thread_id(thread_id: str):
    SESSION_FILE.write_text(json.dumps({"thread_id": thread_id}), encoding="utf-8")


provider = LLMProvider({
    "command": "codex.cmd",
    "args": ["app-server"],
    "prompt_mode": "first_full_then_last",
    "resume_session": True,
    "approval_policy": "never",
    "sandbox": "danger-full-access",
    "cwd": r"C:\Users\yibinjiang\Documents\GitHub\xiaozhi-esp32-server",
    "env": {
        "RUST_LOG": "error",
    },
})

thread_id = load_thread_id()
if thread_id:
    provider._session_map[SESSION_ID] = thread_id
    provider._prompt_mode_state[SESSION_ID] = True

dialogue = [
    {"role": "system", "content": "You are a concise assistant."},
]


def chat_once(user_text: str):
    dialogue.append({"role": "user", "content": user_text})
    response = "".join(provider.response(SESSION_ID, dialogue)).strip()
    if response:
        print(response)
        dialogue.append({"role": "assistant", "content": response})

    thread_id = provider._get_thread_id(SESSION_ID)
    if thread_id:
        save_thread_id(thread_id)


while True:
    user_text = input("You: ").strip()
    if not user_text:
        continue
    if user_text.lower() in {"exit", "quit"}:
        break
    chat_once(user_text)
```
